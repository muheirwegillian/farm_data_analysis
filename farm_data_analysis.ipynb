## Task 1: Load and Explore the Dataset

- Load a dataset (Iris dataset is used here).
- Inspect first few rows.
- Explore structure and missing values.
- Clean if necessary.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Make plots look nice
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

# Load dataset
iris = load_iris(as_frame=True)
df = iris.frame
print("Dataset loaded successfully ")

# Inspect first few rows
print("\nFirst 5 rows of the dataset:")
print(df.head())

# Explore structure
print("\nDataset Info:")
print(df.info())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())
```

## Task 2: Basic Data Analysis

- Compute statistics of numerical columns.
- Perform groupings by categorical column.
- Identify patterns or interesting findings.

```python
# Basic statistics
print("\nBasic statistics of numerical columns:")
print(df.describe())

# Group by target (species)
grouped = df.groupby("target").mean()
print("\nMean values grouped by species (target):")
print(grouped)

# Observations
print("\nObservation: Setosa has the smallest petal length, Virginica has the largest.")
```

## Task 3: Data Visualization

- Create different visualizations with Matplotlib & Seaborn:
  1. Line Chart
  2. Bar Chart
  3. Histogram
  4. Scatter Plot

```python
# 3.1 Line chart
plt.plot(df.index, df["sepal length (cm)"], label="Sepal Length")
plt.title("Line Chart: Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.show()

# 3.2 Bar chart
species_means = df.groupby("target")["petal length (cm)"].mean()
species_means.plot(kind="bar", color=["skyblue", "orange", "green"])
plt.title("Bar Chart: Avg Petal Length per Species")
plt.xlabel("Species (0=Setosa, 1=Versicolor, 2=Virginica)")
plt.ylabel("Petal Length (cm)")
plt.show()

# 3.3 Histogram
df["sepal width (cm)"].plot(kind="hist", bins=15, color="purple", alpha=0.7)
plt.title("Histogram: Sepal Width Distribution")
plt.xlabel("Sepal Width (cm)")
plt.show()

# 3.4 Scatter plot
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="target", data=df, palette="deep")
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()
```
